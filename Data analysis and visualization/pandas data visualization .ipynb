```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df1= pd.DataFrame(np.random.randn(1000, 4), columns=["A", "B", "C", "D"])
```

```python
df1.head()
```

```python
df1["A"].hist(bins=30, grid=False, ec="black", figsize=(4,2))
```

```python
df1["A"].plot(kind="hist", ec="b", bins=30, figsize=(4,3))
```
