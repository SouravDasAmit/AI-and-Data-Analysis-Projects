```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x= np.linspace(0, 5, 11)
y= x**2
```

```python
x
```

```python
y
```

```python
#functional method 
plt.plot(x, y)
plt.show()
```

```python
#set label name
plt.figure(figsize=(4,3))
plt.plot(x, y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("title")
```

```python
#draw subplot
#frist perametter is how many row and secound perametter is fow coloum and last one figure number
plt.subplot(1 , 2, 1) 
plt.plot(x, y, "r") #last perameter is for plot line colour
plt.subplot(1, 2, 2)
plt.plot(y, x, "g")
```

```python
#object oriyended
fig= plt.figure()
axes= fig.add_axes([0.1, 0.1, 0.5, 0.5])
axes.plot(x, y)
```

```python
fig= plt.figure()
axes= fig.add_axes([.05, .05, .5, .5])
axes.plot(x, y)
```

```python
fig= plt.figure()
axes1= fig.add_axes([0.0, 0.0, 0.5, 0.5])
axes1.set_title("figure1")
axes2= fig.add_axes([0.15, 0.15, 0.3, 0.3])
axes2.set_title("figure2")
axes1.plot(x, y)
axes2.plot(y, x)
```

```python
fig,axes= plt.subplots(nrows=1, ncols= 2)
axes[0].plot(x, y)
axes[0].set_title("first graph")
axes[1].plot(y, x)
axes[1].set_title("second graph")
```

```python
#figure size and DPI(dots per inch or pixel per inch)
fig= plt.figure(figsize=(2, 2), dpi=100)
axes= fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
#figure size and DPI(dots per inch or pixel per inch)
fig= plt.figure(figsize=(4, 2), dpi=100)
axes= fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
fig,axes= plt.subplots(nrows=2, ncols=1, figsize=(5, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)
axes[0].set_title("graph 1")
axes[1].set_title("graph 2")
#here two picture are overlapping eatch other so solve this problem
plt.tight_layout()
```

```python
#add legend in a figure
fig= plt.figure()
ax= fig.add_axes([0, 0, .5, .5])
ax.plot(x, x**2, label="x squre")
ax.plot(x, x**3, label="x cube")
ax.legend(loc= 3)
#ax.legend(loc=(.5, .5))
```

```python
# give colour in line you can use hex code 
fig= plt.figure()
ax= fig.add_axes([0, 0, .5, .5])
ax.plot(x, y, color="red")
```

```python
fig= plt.figure()
ax= fig.add_axes([0, 0, .5, .5])
ax.plot(x, y, color="green", linewidth= 3)
```

```python
fig= plt.figure()
ax= fig.add_axes([0, 0, .5, .5])
ax.plot(x, y, color="green", linewidth= 3, alpha=.3)
```

```python
#you can use ls for linestyle
fig= plt.figure()
ax= fig.add_axes([0, 0, .5, .5])
ax.plot(x, y, color="green", lw= 2, linestyle="--")
```

```python
#you can use marker to determine x and y value
fig= plt.figure()
ax= fig.add_axes([0, 0, .5, .5])
ax.plot(x, y, color="green", lw= 3, marker="o")
```

```python
#you can modefy marker color and size
#markersize= ms, markerfacecolor= mfc, markeredgewidth= mew, markeredgecolor= mec
fig= plt.figure()
ax= fig.add_axes([0, 0, .5, .5])
ax.plot(x, y, color="green", lw= 3, marker="o", ms=10, mfc="red", mew=2, mec= "black")
```

```python
fig= plt.figure()
ax= fig.add_axes([0, 0,.51, .5 ])
ax.plot(x, y, color="green", lw= 3, ls="--")
ax.set_xlim([0, 4])
ax.set_ylim([0, 8])
```
